Remove outliers very high price

```python
q25 = np.percentile(dta['price'],25)
q75 = np.percentile(dta['price'],75)

sigma =  0.74 * (q75 - q25)
mean = dta['price'].mean()
dta = dta[(dta['price'] <= mean+5*sigma) & (dta['price'] >= mean-5*sigma)]
```

### 5. Conduct tests to determine which variables are worth exploring

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import *
from sklearn.feature_selection import *

selection = SelectFromModel(RandomForestClassifier(random_state=1234))
```

Use randomforest to select the varalbes to use in the ols

```python
# Select numerical variablres and categorical with dummy variables
y  = dta['price']
X  = dta.drop(['price'],axis=1)
X = dta[['host_listings_count','host_total_listings_count','accommodates',
         'bedrooms','beds','minimum_nights','maximum_nights','number_of_reviews','number_of_reviews_ltm'
        ,'number_of_reviews_l30d','review_scores_rating','review_scores_accuracy',
        'review_scores_cleanliness','review_scores_checkin','review_scores_communication','review_scores_location'
        ,'review_scores_value','calculated_host_listings_count','calculated_host_listings_count_entire_homes',
        'calculated_host_listings_count_private_rooms','calculated_host_listings_count_shared_rooms',
         'reviews_per_month','host_is_superhost','host_has_profile_pic','neighbourhood_cleansed',
         'host_identity_verified','minimum_minimum_nights','maximum_minimum_nights','minimum_maximum_nights',
        'maximum_maximum_nights','minimum_nights_avg_ntm','minimum_nights_avg_ntm','maximum_nights_avg_ntm',
        'availability_30','availability_60','availability_90','availability_365',]]
selection.fit(X,y)
```

```python
features = X.columns[(selection.get_support())]
print(len(features))
features
```

### 6. Build models using OLS regression to predict listing price

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    dta[features],
    dta[['price']],
    test_size=0.25,
    random_state=1234)
```

```python
import statsmodels.api as sm
```

```python
mod = sm.OLS(y_train, X_train)  
res = mod.fit()      
print (res.summary())
```

Some varaibels are not significant train the model again with important varaibels

```python
mod = sm.OLS(y_train, X_train[['bedrooms', 'beds', 'maximum_nights', 'number_of_reviews',
       'number_of_reviews_ltm', 'review_scores_rating',
       'review_scores_cleanliness', 'review_scores_value', 'reviews_per_month',
       'host_identity_verified', 'availability_365']])  
res = mod.fit()      
print (res.summary())
```

```python
res.params
```

```python
pred = res.predict(X_test[['bedrooms', 'beds', 'maximum_nights', 'number_of_reviews',
       'number_of_reviews_ltm', 'review_scores_rating',
       'review_scores_cleanliness', 'review_scores_value', 'reviews_per_month',
       'host_identity_verified', 'availability_365']])
```
